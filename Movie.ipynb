```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Movies.csv")
df
```

```python
genre_counts = df['Genres'].str.split(',').explode().value_counts()
print('Contagem de Filmes por Gênero:')
print(genre_counts)
```

```python
# Média
mean_rating = df['IMDB Rating'].mean()

# Mediana
median_rating = df['IMDB Rating'].median()

# Desvio Padrão
std_rating = df['IMDB Rating'].std()

print(f'Média das Notas IMDB: {mean_rating}')
print(f'Mediana das Notas IMDB: {median_rating}')
print(f'Desvio Padrão das Notas IMDB: {std_rating}')
```

```python
# Verificar as primeiras linhas do dataframe
print(df.head())

# Histograma das notas dos filmes
plt.figure(figsize=(10, 6))
sns.histplot(df['IMDB Rating'], bins=20, kde=True)
plt.title('Distribuição das Notas dos Filmes no IMDB')
plt.xlabel('Nota')
plt.ylabel('Frequência')
plt.show()

# Estatísticas básicas
media = df['IMDB Rating'].mean()
mediana = df['IMDB Rating'].median()
moda = df['IMDB Rating'].mode()[0]

print(f'Média das notas: {media}')
print(f'Mediana das notas: {mediana}')
print(f'Moda das notas: {moda}')
```

```python
numeric_columns = ['Release Year', 'Length in Min', 'IMDB Rating', 'Rating Count']
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())

# Gráfico de dispersão para Duração dos Filmes vs Notas
plt.figure(figsize=(10, 6))
plt.scatter(df['Length in Min'], df['IMDB Rating'])
plt.title('Duração dos Filmes vs Notas')
plt.xlabel('Duração dos Filmes (Minutos)')
plt.ylabel('Nota IMDB')
plt.show()

# Gráfico de dispersão para Quantidade de Avaliações vs Notas
plt.figure(figsize=(10, 6))
plt.scatter(df['Rating Count'], df['IMDB Rating'])
plt.title('Quantidade de Avaliações vs Notas')
plt.xlabel('Quantidade de Avaliações')
plt.ylabel('Nota IMDB')
plt.show()
```

```python
longest_movie = df.loc[df['Length in Min'].idxmax()]
shortest_movie = df.loc[df['Length in Min'].idxmin()]

print(f'Filme com a Maior Duração: {longest_movie["Title"]} - {longest_movie["Length in Min"]} minutos')
print(f'Filme com a Menor Duração: {shortest_movie["Title"]} - {shortest_movie["Length in Min"]} minutos')
```

```python
# Selecionar apenas as colunas numéricas
numeric_columns = ['Release Year', 'Length in Min', 'IMDB Rating', 'Rating Count']
df_numeric = df[numeric_columns].copy()

# Verificar valores nulos e preencher com a média apenas nas colunas numéricas
df_numeric.fillna(df_numeric.mean(), inplace=True)

# Calcular a matriz de correlação
correlation_matrix = df_numeric.corr()

print('Matriz de Correlação:')
print(correlation_matrix)
```

```python
# Verificar valores nulos e preencher com a média apenas nas colunas numéricas
numeric_columns = ['Release Year', 'Length in Min', 'IMDB Rating', 'Rating Count']
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())

# Calcular estatísticas descritivas das notas IMDB
rating_stats = df['IMDB Rating'].describe()

print('Estatísticas Descritivas das Notas IMDB:')
print(rating_stats)

# Plotar histograma das notas IMDB
plt.figure(figsize=(10, 6))
plt.hist(df['IMDB Rating'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribuição das Notas IMDB')
plt.xlabel('Nota IMDB')
plt.ylabel('Frequência')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
